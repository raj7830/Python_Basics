3. a=(1, −3, 5, 6)T

 b = (7, 4, −9, 0)T

   (a) Using a suitable data structure store a and b.

   (b) Find the inner product of a and b.


```python
a=[1,-3,5,6]
b=[7,4,-9,0]

inner_product = sum(a[i] * b[i] for i in range(len(a)))

print(inner_product)
```

5. (a) Generate 1000 random numbers from [0,1) and store the first 500 in vector x
and remaining in vector y.

 (b) Find z = x ∗ (x + y). Plot scatter plot for representing z. Set axis labels and
figure title.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

random_n = np.random.random(1000)

x = random_n[0:500]
y = random_n[500:]
z = x*(x+y)

ax = plt.axes(projection = '3d')
ax.scatter3D(x,y,z)
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```
