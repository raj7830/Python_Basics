1. Consider x = [1, 2, 3, 3, 4, 5, 5, 5, 6, 8, 0, 10]

(a) Write a program to find the frequency (number of occurrences) of each element
in x.

```python


list_x=[1,2,3,3,4,5,5,5,6,8,0,10]
freq={}

for item in list_x:
    if item in freq:
        freq[item]=item+1
    else:
        freq[item]=1
print(freq)
```

5. Consider S=[a,o,h,j,e,j].

Write a program to perform the following task: Read
each alphabet from S. If the alphabet is a vowel: print ”The alphabet is a vowel”.
Construct a string by concatenating the alphabet three times. Else: print ”The
alphabet is a consonant”. Construct a string by concatenating the alphabet five
times.

```python


def str_ops():
    vowel=['a','e','i','o','u','A','E','I','O','U']
    S=['a','o','h','j','e','j']

    for element in S:
        if element in vowel:
            print("The element is Vowel")
            con_str=element*3
        else:
            print("The element is a Consonant")
            con_str=element*5
        print("con_str:",con_str)
        print()
str_ops()
```

```python


#construct A and B
A=[[3,-1,4],
   [1,5,6],
   [1,0,2]]

B=[[2,-1,1],
   [3,2,-6],
   [4,-5,2]]

print(A)
print(B)
```

```python
#(b) A+B

Addition =[]

for i in range(len(A)):
    row = []
    for j in range(len(A)):
       element_add = A[i][j] + B[i][j]
       row.append(element_add)
    Addition.append(row)
for row in Addition:
    print(row)
```

```python
#8(c) A*B

multiplication = []

for i in range(len(A)):
  row = []
  for j in range(len(B[0])):
    sum = 0
    for k in range(len(A[0])):
      sum += A[i][j] * B[k][j]
    row.append(sum)
  multiplication.append(row)

for row in multiplication:
  print(row)
```
