4. Plot normal distribution with mean 0 and standard deviation 1.

```python


import matplotlib.pyplot as plt
import math
def normal(z,std,mean):
  n = []

  for i in z:
    pi = math.pi

    v = (2*pi*std**2)**(0.5)
    w = (i-mean)/std
    f = (1/v)*np.exp((-0.5)*(w)**2)
    n.append(f)

  return n
```

```python

z = np.linspace(-10,10,1000)
Normal_distribution = normal(z,1,0)

plt.plot(Normal_distribution)
plt.title("Normal Distribution")
```

6. Generate 200 data points that satisfy the equation 3x + 4y − 3 = 0. Plot the
hyperplane and mark the selected points.

```python
x = np.linspace(0,50,200)

y = []
def data(x):

  for i in range(len(x)):
    temp = (3-3*x[i])/4
    y.append(temp)
  return y
```

```python
plane = data(x)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(plane)
```

```python
x = np.linspace(0,1,1000)
y = np.linspace(0,5,1000)
z = []

for i in range(len(x)):
  temp = (5*x[i]-2*y[i])/6
  z.append(temp)

print(z)
```
